```python

def classify(petal_length, petal_width):
  if petal_length <= 2.0:
    return "setosa"
  elif petal_length <= 2.5 and petal_width <= 1.5:
    return "setosa"
  elif petal_length <= 3.0 and petal_width <= 1.0:
    return "setosa"
  elif petal_length <= 3.3 and petal_width <= 2.0:
    return "versicolor"
  else:
    return "virginica"
examples = [
  {"petal_length": 5.1, "petal_width": 2.4},
  {"petal_length": 4.0, "petal_width": 1.0},
  {"petal_length": 1.4, "petal_width": 0.2},
  {"petal_length": 6.3, "petal_width": 1.8},
  {"petal_length": 1.5, "petal_width": 0.2},
]
actual_classes = ["virginica", "versicolor", "setosa", "virginica", "setosa"]

correct_classifications = 0
for example, actual_class in zip(examples, actual_classes):
  predicted_class = classify(example["petal_length"], example["petal_width"])
  if predicted_class == actual_class:
    correct_classifications += 1

accuracy = correct_classifications / len(examples) * 100
print(f"Accuracy: {accuracy:.2f}%")
```

```python
#Exercise 3.1
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
X = np.array([[1.0], [2.0], [3.0]])
y = np.array([0, 1, 0])
max_depth_values = [2, 5, 10]
plt.figure(figsize=(15, 10))

for i, max_depth in enumerate(max_depth_values, 1):
    clf = tree.DecisionTreeClassifier(random_state=0, max_depth=max_depth)
    pair = [0]
    clf.fit(X[:, pair], y)
    plt.subplot(2, 2, i)
    plt.scatter(X[:, pair[0]], y, edgecolors='k', cmap=plt.cm.Paired)
    plot_step = 0.01
    x_min, x_max = X[:, pair[0]].min() - 1, X[:, pair[0]].max() + 1
    xx = np.arange(x_min, x_max, plot_step)
    Z = clf.predict(xx.reshape(-1, 1))
    plt.step(xx, Z, where='mid', color='r', lw=2)
    plt.xlabel('Feature')
    plt.ylabel('Target')
    plt.title(f'Decision Boundary (max_depth={max_depth})')

plt.tight_layout()
plt.show()
```

```python
#Exercise 3.2
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
iris = datasets.load_iris()
X = iris.data[:, [2, 3]]
y = iris.target
clf = DecisionTreeClassifier(max_depth=3, random_state=0)
clf.fit(X, y)
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.Paired)
plot_step = 0.01
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step), np.arange(y_min, y_max, plot_step))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Paired)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Decision Boundaries of Decision Tree (max_depth=3)')
plt.show()
```

```python
#Exercise 4
def predictSpecies(sepal_width, sepal_length, petal_width, petal_length) -> str:
    """
    Predict the species of the plant using a decision tree.

    Parameters:
    - sepal_width: Sepal width of the flower
    - sepal_length: Sepal length of the flower
    - petal_width: Petal width of the flower
    - petal_length: Petal length of the flower

    Returns:
    - str: Predicted species (e.g., "setosa", "versicolor", "virginica")
    """
    if petal_length <= 2.45:
        if petal_width <= 1.75:
            return "setosa"
        else:
            return "versicolor"
    else:
        if petal_width <= 1.65:
            if sepal_length <= 4.95:
                return "versicolor"
            else:
                return "virginica"
        else:
            if sepal_length <= 5.95:
                return "versicolor"
            else:
                return "virginica"
predicted_species = predictSpecies(3.0, 5.0, 1.5, 2.0)
print(f"Predicted Species: {predicted_species}")
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
data = {
    'sepal_length': [5.8, 6.1, 5.8, 4.4, 7.2, 4.9, 4.9, 5.9, 5.0, 5.4],
    'sepal_width': [2.7, 2.6, 2.8, 3.2, 3.6, 2.4, 3.0, 3.2, 3.0, 3.0],
    'petal_length': [3.9, 5.6, 5.1, 1.3, 6.1, 3.3, 1.4, 4.8, 1.6, 4.5],
    'petal_width': [1.2, 1.4, 2.4, 0.2, 2.5, 1.0, 0.2, 1.8, 0.2, 1.5],
    'Class': [1, 2, 2, 0, 2, 1, 0, 1, 0, 1],
    'ClassName': ['versicolor', 'virginica', 'virginica', 'setosa', 'virginica',
                  'versicolor', 'setosa', 'versicolor', 'setosa', 'versicolor']
}

df = pd.DataFrame(data)
X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = df['Class']
clf = DecisionTreeClassifier(random_state=0, max_depth=3)
clf.fit(X, y)
def predict_species(sepal_length, sepal_width, petal_length, petal_width) -> str:
    prediction = clf.predict([[sepal_length, sepal_width, petal_length, petal_width]])[0]
    class_names = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
    return class_names[prediction]
sample_instance = (5.4, 3.0, 4.5, 1.5)
predicted_species = predict_species(*sample_instance)
print(f"Predicted Species: {predicted_species}")
```

```python
import pandas as pd
import numpy as np
data = {
    'sepal_length': [5.8, 6.1, 5.8, 4.4, 7.2, 4.9, 4.9, 5.9, 5.0, 5.4],
    'sepal_width': [2.7, 2.6, 2.8, 3.2, 3.6, 2.4, 3.0, 3.2, 3.0, 3.0],
    'petal_length': [3.9, 5.6, 5.1, 1.3, 6.1, 3.3, 1.4, 4.8, 1.6, 4.5],
    'petal_width': [1.2, 1.4, 2.4, 0.2, 2.5, 1.0, 0.2, 1.8, 0.2, 1.5],
    'Class': [1, 2, 2, 0, 2, 1, 0, 1, 0, 1],
    'ClassName': ['versicolor', 'virginica', 'virginica', 'setosa', 'virginica',
                  'versicolor', 'setosa', 'versicolor', 'setosa', 'versicolor']
}

df = pd.DataFrame(data)
class_counts = df['Class'].value_counts()
total_samples = len(df)
class_probabilities = class_counts / total_samples
entropy = -np.sum(class_probabilities * np.log2(class_probabilities))

print(f"Entropy: {entropy}")
```

```python
#Exercise 6
import pandas as pd
import numpy as np
from sklearn.metrics import mutual_info_score
total_entropy = -np.sum((df['Class'].value_counts() / len(df)) * np.log2(df['Class'].value_counts() / len(df)))
def calculate_information_gain(feature, threshold):
    left_group = df[df[feature] <= threshold]['Class']
    right_group = df[df[feature] > threshold]['Class']
    left_entropy = -np.sum((left_group.value_counts() / len(left_group)) * np.log2(left_group.value_counts() / len(left_group)))
    right_entropy = -np.sum((right_group.value_counts() / len(right_group)) * np.log2(right_group.value_counts() / len(right_group)))
    information_gain = total_entropy - (len(left_group) / len(df) * left_entropy + len(right_group) / len(df) * right_entropy)

    return information_gain
best_information_gain = 0
best_split = None
for feature in df.columns[:-2]:
    thresholds = df[feature].unique()
    for threshold in thresholds:
        current_information_gain = calculate_information_gain(feature, threshold)
        if current_information_gain > best_information_gain:
            best_information_gain = current_information_gain
            best_split = (feature, threshold)

print(f"Best Split: {best_split}, Information Gain: {best_information_gain}")
```
